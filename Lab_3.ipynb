```python
# Step 1: Import necessary libraries
import pandas as pd
from sklearn.impute import KNNImputer
from sklearn.preprocessing import LabelEncoder

# Step 2: Load the dataset
data = pd.read_csv("supermarket_sales2.csv")

df = pd.DataFrame(data)

# Step 3: Preprocess the data
# Convert categorical variables to numerical values
label_encoders = {}
for column in df.select_dtypes(include=['object']).columns:
    if column != 'Invoice ID' and column != 'Date' and column != 'Time':
        le = LabelEncoder()
        df[column] = le.fit_transform(df[column])
        label_encoders[column] = le

# Step 4: Apply KNN Imputer
imputer = KNNImputer(n_neighbors=2)
df_imputed = pd.DataFrame(imputer.fit_transform(df.select_dtypes(include=['number'])), columns=df.select_dtypes(include=['number']).columns)

# Step 5: Post-process the data
# Convert numerical values back to categorical if needed
for column, le in label_encoders.items():
    df_imputed[column] = le.inverse_transform(df_imputed[column].astype(int))

# Combine imputed numerical data with original non-numerical data
df_final = df_imputed.join(df[['Invoice ID', 'Date', 'Time']])

print(df_final)
```
